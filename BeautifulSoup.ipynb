```python
import bs4
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cryptocurrencies'
data = urlopen(url)
```

```python
data
```

```python
html = data.read()
data.close()
```

```python
from bs4 import BeautifulSoup as soup
```

```python
soup_crypto = soup(html, 'html.parser')
```

```python
tables = soup_crypto.find_all('table', {'class':'wikitable'})
```

```python
tables[0]
```

```python
rows = tables[0].find_all('tr')
```

```python
len(rows)
```

```python
matrix_data = []
for row in rows:
  row_data=[]
  for cell in row:
    if(len(cell.text)>1):
    # data contains ','. 
    # Thus pandas wont be able to read csv file
    # just by row_data.append(cell.text[:-1])
    # hence we use
      row_data.append(cell.text[:-1].replace(',', '-'))
  matrix_data.append(row_data)
```

```python
matrix_data
```

```python
csv_file = 'active_crypto.csv'
with open(csv_file, 'w') as f:
  for row in matrix_data:
    row_csv = ','.join(row)
    row_csv += '\n'
    # print((row_csv))
    f.write(row_csv)
```
